```python
#Import libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv("csse_covid_19_time_series/time_series_covid19_deaths_global.csv")
df.head()
```

The data needs to be transformed for it to be useful. 

```python
# Transform the data using padas melt
df_melt = pd.melt(df, id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'], var_name='date', value_name='numdeaths')
df_melt.tail(10)
```

```python
df_melt[df_melt['Country/Region']=="Algeria"].plot(x='date', y='numdeaths', 
                                                   linewidth=2, figsize=(20,5), rot=45);
plt.title("Number of deaths", fontsize=14); 
plt.xlabel('Day', fontsize=14)
```

```python
# Export transformed data as CSV
df_melt.to_csv("transformed-data/transformed-global-covid19-deaths.csv", index=False)
```
